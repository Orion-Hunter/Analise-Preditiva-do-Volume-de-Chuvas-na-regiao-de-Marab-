```python
def deslocar(delta: int, X, y):
    vx = X[0:(delta+1)]
    vy = y[(delta+1):delta*2]
    return vx, vy, stats.spearmanr(vx, vy)
```

```python
def defasar(delta, X, Y):
    l = {'Área de Precipitação': Y.columns[1:]}
    df = {}
    for i in data.columns[1:]:
        l[str(i)+"("+str(delta)+")"]  = []
        for j in prm.columns[1:]:
            x_values = X[i][Y[delta:].index].values
            y_values = Y[delta:][str(j)].values
            d = {str(i)+"("+str(delta)+")": x_values, "PRM": y_values}
            cor = stats.spearmanr(x_values, y_values)[0]
            l[str(i)+"("+str(delta)+")"].append(float(cor))
            df[str(i)+"("+str(delta)+")"] = d
    return df, l  
```

```python
frames = {}
dfs = []
for r in range(0, 13):
    datas, lists =  defasar(r, data, prm)
    dfs.append(datas)
    frames[r] = pd.DataFrame(lists)
```

```python
#mask = frames[0]['TSM-AN1(0)'].between(-0.95,-0.7) | frames[0]['TSM-AN1(0)'].between(0.7,0.95)
filtered50 = {}
filtered55 = {}
filtered60 = {}
filtered65 = {}


def filter(data, filter_rate):
    filtered = {}
    for i in data.columns[1:]:
        tmp = []
        for j in data[i]:
            if (j <= -filter_rate ) or (j >= filter_rate):
                tmp.append(j)
            else:
                tmp.append(str('\x20'))
        filtered[i] = tmp
        tmp = []
    return filtered

for key in frames.keys():
    filtered50[key] = filter(frames[key], 0.5)
    filtered55[key] = filter(frames[key], 0.55)
    filtered60[key] = filter(frames[key], 0.6)
    filtered65[key] = filter(frames[key], 0.65)
```

```python
pd.DataFrame(filtered50[0])
```

```python
variaveis_por_filtro = {">= 50%": []}
variaveis = []
dados = []
i = 0


def verify_var(data,k, j):
    i = 0
    if ' ' not in data[k][j]:
        return j
    return False
    
    
    
def generate_filter_info(data: pd.DataFrame, range_filter):
    info = {"Quantidade(Variáveis sem valores fora do filtro)": [], "Variaveis": [], "Filtro": []}
    for k in data.keys():
        i = 0
        var  = []
        for j in data[k].keys():
            if verify_var(data,k, j):
                i+=1
                var.append(j)
        info['Quantidade(Variáveis sem valores fora do filtro)'].append(i)
        info['Variaveis'].append(var)
        info['Filtro'].append(range_filter)
    return info    
    
pd.DataFrame(generate_filter_info(filtered65, "65%"))
```

```python
excel_writer1 = pd.ExcelWriter('filtros50.xlsx', engine='xlsxwriter')
excel_writer2 = pd.ExcelWriter('filtros55.xlsx', engine='xlsxwriter')
excel_writer3 = pd.ExcelWriter('filtros60.xlsx', engine='xlsxwriter')
excel_writer4 = pd.ExcelWriter('filtros65.xlsx', engine='xlsxwriter')

info1 = generate_filter_info(filtered50, "50%")
info2 = generate_filter_info(filtered55, "55%")
info3 = generate_filter_info(filtered60, "60%")
info4 = generate_filter_info(filtered65, "65%")

for r in frames.keys():
    pd.DataFrame(filtered50[r]).to_excel(excel_writer1, sheet_name='Defasagem de '+str(r)+' meses')
    pd.DataFrame(info1).to_excel(excel_writer1, sheet_name="Informações sobre as variáveis")
    
    pd.DataFrame(filtered55[r]).to_excel(excel_writer2, sheet_name='Defasagem de '+str(r)+' meses')
    pd.DataFrame(info2).to_excel(excel_writer2, sheet_name="Informações sobre as variáveis")  
    
    pd.DataFrame(filtered60[r]).to_excel(excel_writer3, sheet_name='Defasagem de '+str(r)+' meses')
    pd.DataFrame(info3).to_excel(excel_writer3, sheet_name="Informações sobre as variáveis")
    
    pd.DataFrame(filtered65[r]).to_excel(excel_writer4, sheet_name='Defasagem de '+str(r)+' meses')
    pd.DataFrame(info4).to_excel(excel_writer4, sheet_name="Informações sobre as variáveis")
    
excel_writer1.save()
excel_writer2.save()
excel_writer3.save()
excel_writer4.save()
```

```python
from matplotlib.backends.backend_pdf import PdfPages



def plot(data):
    fig = plt.figure(figsize=(15,10))
    g = plt.boxplot(data)
    return fig


pp = PdfPages('Visualizacao65%.pdf')
for r in range(0,13):
    df = pd.DataFrame(filtered65[r])
    columns = pd.DataFrame(info4).loc[r,'Variaveis']
    fig = plot(df[columns])
    plt.title("Defasagem de "+str(r)+" meses")
    plt.xlabel("Variáveis")
    plt.ylabel("Valores de Correlação")
    plt.legend(columns,bbox_to_anchor=(0.14, 0.82, 1., .102))
    pp.savefig(fig)

pp.close()
```
