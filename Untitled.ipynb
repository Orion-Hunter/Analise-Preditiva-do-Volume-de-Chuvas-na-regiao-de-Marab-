```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = ak.StructuredDataClassifier(
    multi_label=True,
    project_name="modelo",
    max_trials=150,
    directory="output/model",
    objective="val_accuracy",
    overwrite=True)
```

```python
model.fit(X_train, y_train)
```
